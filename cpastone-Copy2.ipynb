#LOAD Latitude and longitude

```python
# define a function to get coordinates
def get_latlng(neighborhood):
    # initialize your variable to None
    lat_lng_coords = None
    # loop until you get the coordinates
    while(lat_lng_coords is None):
        g = geocoder.arcgis('{}, Denver, Colorado'.format(neighborhood))
        lat_lng_coords = g.latlng
    return lat_lng_coords

coords = [ get_latlng(neighborhood) for neighborhood in Denver_df["Neighborhood"].tolist() ]

coords
```

#Lets analyze these areas. We need to perfom onehot encoding on categorical values

```python
# one hot encoding
Denver_onehot = pd.get_dummies(venues_df[['VenueCategory']], prefix="", prefix_sep="")

# add postal, borough and neighborhood column back to dataframe
Denver_onehot['Neighborhoods'] = venues_df['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [Denver_onehot.columns[-1]] + list(Denver_onehot.columns[:-1])
Denver_onehot = Denver_onehot[fixed_columns]

print(Denver_onehot.shape)
Denver_onehot.head()
```

#Now we need to group by neighborhood by taking the mean of the frequency of occurrence of each category

```python
Denver_grouped = Denver_onehot.groupby(["Neighborhoods"]).mean().reset_index()

print(Denver_grouped.shape)
Denver_grouped
```

```python

len(Denver_grouped[Denver_grouped["Asian Restaurant"] > 0])
```

#Now that we have grouped the rows. Lets take somes top venues for each Postalcod based on the values 

```python
Denver_AR = Denver_grouped[["Neighborhoods","Asian Restaurant"]]
Denver_AR
```

#We can now go down to Clustering with the top venues

```python
# set number of clusters
kclusters = 5

Denver_grouped_clustering = Denver_AR.drop(["Neighborhoods"], 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Denver_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood.
Denver_merged = Denver_AR.copy()


# add clustering labels
Denver_merged["Cluster Labels"] = kmeans.labels_

Denver_merged.rename(columns={"Neighborhoods": "Neighborhood"}, inplace=True)
#_merged.head()

# merge denver_grouped with denver_data to add latitude/longitude for each neighborhood
Denver_merged = Denver_merged.join(Denver_df.set_index("Neighborhood"), on="Neighborhood")

print(Denver_merged.shape)
Denver_merged.head()
```

```python
# sort the results by Cluster Labels
Denver_merged.sort_values(["Cluster Labels"], inplace=True)
Denver_merged
```

#Visualization of the cluster

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon,poi, cluster in zip(Denver_merged['Latitude'], Denver_merged['Longitude'], Denver_merged['Neighborhood'], Denver_merged['Cluster Labels']):
    label = folium.Popup('{} - Cluster {}'.format(poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters



```
